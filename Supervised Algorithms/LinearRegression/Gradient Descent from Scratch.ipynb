```python
import numpy as np 
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=50, random_state=15)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=13)
```

```python
plt.scatter(X_train,y_train)
plt.show()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

# Gradient Descent from Class

```python
class GDRegressor:
    
    def __init__(self, learning_rate , epoch):
        self.m = 100
        self.b = -120
        self.lr = learning_rate
        self.epochs = epoch
    
    def fit(self,X,y):
        for i in range(self.epochs):
            loss_slop_at_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            loss_slop_at_m = -2 * np.sum((y - self.m*X.ravel() - self.b)*X.ravel())
            self.b = self.b - self.lr*loss_slop_at_b
            self.m = self.m - self.lr*loss_slop_at_m
            
            y_pred = self.m * X + self.b
            plt.plot(X,y_pred)
            
#         print(loss_slop,self.b)         
#         print(self.m,self.b)

        plt.scatter(X,y)
        plt.show()

    def predict(self,X):
        
        return self.m * X.ravel() + self.b
   
```

```python
gdr = GDRegressor(0.001,100)
```

```python
gdr.fit(X_train,y_train)
```

```python
y_pred1 = gdr.predict(X_test)
```

```python
r2_score(y_test,y_pred1)
```
