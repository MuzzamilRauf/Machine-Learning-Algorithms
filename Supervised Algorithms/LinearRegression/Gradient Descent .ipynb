```python
from sklearn.datasets import make_regression
import numpy as np 
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=50, random_state=13)
```

```python
print(X)
print(y)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y)
plt.show
```

```python
# Let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Let's apply Gradient Descent assuming slop is constant m=71.81
# and Let's assume the sarting value for intercept b=0
y_pred = ((71.81 * X) + 0).reshape(4)
```

```python
y_pred = ((71.81 * X) + 0).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label = 'OLS')
plt.plot(X,y_pred,color='b', label='b=0')
plt.legend()
plt.show()
```

```python
m = 71.81
b = 0

loss_slop = -2 * np.sum(y - m*X.ravel() - b)
loss_slop
```

```python
# lets take Learning rate = 0.1
lr = 0.1
step_size = loss_slop*lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((71.81 *X) + b).reshape(4) 

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label = 'OLS')
plt.plot(X,y_pred1,color='b', label='b = {}'.format(b))
plt.plot(X,y_pred,color='k', label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slop = -2 * np.sum(y - m*X.ravel() - b)
loss_slop
```

```python
step_size = loss_slop*lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred2 = ((71.81 *X) + b).reshape(4) 

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label = 'OLS')
plt.plot(X,y_pred1,color='b', label='b = {}'.format(b))
plt.plot(X,y_pred2,color='g', label='b = {}'.format(b))
plt.plot(X,y_pred,color='k', label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slop = -2 * np.sum(y - m*X.ravel() - b)
loss_slop
```

```python
step_size = loss_slop*lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred3 = ((71.81 *X) + b).reshape(4) 

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label = 'OLS')
plt.plot(X,y_pred1,color='b', label='b = {}'.format(b))
plt.plot(X,y_pred2,color='g', label='b = {}'.format(b))
plt.plot(X,y_pred3,color='w', label='b = {}'.format(b))
plt.plot(X,y_pred,color='k', label='b=0')
plt.legend()
plt.show()
```

```python
m = 71.81
b = -10
lr = 0.1

epochs = 12

for i in range(epochs):
    loss_slop = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slop)
    
    y_pred = (m * X + b)
    
    plt.plot(X,y_pred)
plt.scatter(X,y)
plt.show()
print(b)
```
